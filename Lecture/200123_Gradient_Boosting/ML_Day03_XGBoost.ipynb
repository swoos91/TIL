```python
import xgboost as xgb
from xgboost import plot_importance

import pandas as pd
import numpy as np

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
```

```python
xgb.__version__
```

```python
dataset = load_breast_cancer()
X_features = dataset.data
y_label = dataset.target

X_features.shape   # ( 569, 30 )
y_label  # 2진 분류  ( m(;악성) = 0, b(;양성) = 1 )
```

```python
cancer_df = pd.DataFrame(data = X_features, columns = dataset.feature_names)
cancer_df['target'] = y_label
cancer_df.head()
```

```python
print(dataset.target_names)
print(cancer_df['target'].value_counts())
```

```python
X_train, X_test, y_train, y_test= train_test_split(X_features, y_label, test_size = 0.2, random_state = 42
                                                  )
```

# XGboost

XGboost는 DMatrix 타입의 인수를 받으므로 데이터 구조 변환이 필요

```python
# read in data
dtrain = xgb.DMatrix(data = X_train, label = y_train)
dtest = xgb.DMatrix(data = X_test, label = y_test)

# type(dtrain)    //   cancer_df['target'].values --> array 구조로 return하므로 이를 인수로 넣어줘도 무방

# specify parameters via map

params = {'max_depth':3,                  # depth 조정을 통해 overfitting 방지 
          'eta':0.1,                      # Theta의 약자로, learning rate 의미
          'objective':'binary:logistic',  # 목적 변수의 형태를 적용하는 함수 지정
          'eval_metric':'logloss',        # 오류 성능 평가 지표( logloss == cost function )
          'early_stoppings':100           # 오류 함수의 성능 향상이 나타나지 않는 경우 stop 할 때의 기준 설정
         }
num_rounds = 400  # Boosting 반복횟수 
wlist = [(dtrain, 'train'), (dtest, 'eval')]   # dtest 는 엄밀한 의미에서의 test data가 아닌 validation 으로 봐야 함. 다시 말해서, 모델의 성능을 높이기 위해(loss율을 낮추기 위해) 훈련 데이터 이외에 모의고사용 테스트로써 dtest 데이터 투입
xgb_model = xgb.train(params = params , \
                dtrain = dtrain, num_boost_round = num_rounds, evals = wlist)
```

```python
pred_probs = xgb_model.predict(dtest)
print('predict( ) 수행 결과값을 10개만 표시, 예측 확률 값으로 표시됨')
print(np.round(pred_probs[:10],3))

# 예측 확률이 0.5 보다 크면 1, 그렇지 않으면 0 으로 예측값 결정하여 List 객체인 preds
preds = [ 1 if x > 0.5 else 0 for x in pred_probs ]
print('예측값 10개만 표시:', preds[:10])
```
